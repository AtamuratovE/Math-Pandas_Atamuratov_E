Algebra problem

*Picture yourself riding a horse, on a thrilling quest to hunt down a bear. At the start, the bear is 200 meters away from you. Your horse dashes at a speedy 60 km/h, while the bear moves at 30 km/h. So, how much time will it take for you to catch up to the bear?*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 600, 1000) # start, finish, n points
```

```python
d_bear = 0.5 * t + 200 # we add 200 because the bear started 200 meters ahead of us
```

```python
 d_you =  1 * t # distance you traveled
```

```python

fig, ax = plt.subplots()
plt.title('A Bear Hunt')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 500])
ax.set_ylim([0, 500])
ax.plot(t, d_bear, c='green')
ax.plot(t, d_you, c='brown')
plt.axvline(x=400, color='purple', linestyle='--')
_ = plt.axhline(y=400, color='purple', linestyle='--')
```

**Answer is 400 minutes**
