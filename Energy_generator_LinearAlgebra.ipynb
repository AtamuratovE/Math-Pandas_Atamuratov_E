jill designs panels as hobby.
on April 1st, jill's "Mark 1" design begins generating power: 1kj/day.
on May 1st, her "Mark 11" design begins generating 4 kj of power per day.

- What day is it when jill's Mark 11 design has generated as much total energy as the Mark 1design?

- How much total energy have both generated by that day?

- What would the solutions to (1.) and (2.) be if Mark 11 design generated 1kj of power per day?


**total_energy = kj/day * num_Of_Days**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.linspace(0, 500, 100)
```

```python
Tot_en_Mark1 = T * 1 + 30
```

```python
Tot_en_Mark11 = T * 4
```

```python
fig, ax = plt.subplots()
plt.title('Energy predictor')
plt.xlabel('time (in days)')
plt.ylabel('energy (in kj)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(T, Tot_en_Mark11, c='green')
ax.plot(T, Tot_en_Mark1, c='brown')
plt.axvline(x=10, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```

 - 1. **May 10st**
 
 - 2. **40 kj - each one, 80 total**
 
 - 3. **1. no sol, 2. Mark1 - 40 kj, Mark11 - 10 kj**
