# $\lim_{x \to 0} f(\frac{x^2 - 1}{x - 1})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fxn(x):
    y = (x**2 - 1)/(x - 1)
    return y
```

```python
x = np.linspace(-15, 15, 100)
y = fxn(x)
```

```python
fxn(2)
```

```python
fxn(-2)
```

#### $\lim_{x \to -2} f(\frac{-2^2 - 1}{-2 - 1}) = -1.0 $

```python
fxn(1) # here showing this error message because we encountered devision by zero 
```

# $\lim_{x \to -5} \frac{x^2 - 25}{x + 5} $

```python
def fx(x):
    y = ((x**2) - 25)/(x + 5)
    return y
```

```python
x = np.linspace(-25, 25, 100)
y = fx(x)
```

```python
fig, ax = plt.subplots()
plt.xlim(-10, 0)
plt.ylim(-20, 0)
_=ax.plot(x,y)
```

Let's try approaching from 0 and -10

```python
fx(0)
```

```python
fx(-10)
```

let's go closer

```python
fx(-2)
```

```python
fx(-8)
```

```python
fig, ax = plt.subplots()
plt.xlim(-8, -2)
plt.ylim(-20, 0)
_=ax.plot(x,y)
```

```python
fx(-4)
```

```python
fx(-6)
```

```python
fig, ax = plt.subplots()
plt.xlim(-6, -4)
plt.ylim(-13, -7.5)
_=ax.plot(x,y)
```

As you can see as much we close to -5 as much it get's closer to -10

```python
fx(-4.999)
```

```python
fx(-5.001)
```

```python
fig, ax = plt.subplots()
plt.xlim(-5.001, -4.999)
plt.ylim(-13, -7.5)
_=ax.plot(x,y)
```

as we can see here values of y is -10 as x approaches -5 from both left and right hand side 
- **The limit of the expression as x approaches -5 is therefore equal to -10.**

# $\lim_{x \to 4} \frac{x^2 - 2x - 8}{x - 4} $

```python
def fx1(x1):
    y1 = ((x1**2) - 2*x1 - 8)/(x1-4)
    return y1
```

```python
x1 = np.linspace(-25, 25, 1000)
y1 = fx1(x1)
```

```python
fig, ax = plt.subplots()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
_=ax.plot(x1,y1)
```

**Zoom 1:** 

```python
fx1(0)
```

```python
fx1(8)
```

```python
fig, ax = plt.subplots()
plt.xlim(0, 8)
plt.ylim(2, 10)
_=ax.plot(x1,y1)
```

It is continious from 1.5 till 8 that is what we need. because 4 is within that range.

let's thoom a bit more

**Zoom 2:** 

```python
fx1(2)
```

```python
fx1(6)
```

```python
fig, ax = plt.subplots()
plt.xlim(2, 6)
plt.ylim(3, 9)
_=ax.plot(x1,y1)
```

**Zoom 3:** 

```python
fx1(3.9)
```

```python
fx1(4.1)
```

```python
fig, ax = plt.subplots()
plt.xlim(3.9, 4.1)
plt.ylim(3, 9)
_=ax.plot(x1,y1)
```

As you can see it's as x approaching 4 y is getting closer and closer to 6 now we zoom last time.

**Zoom 3:** 

```python
fx1(3.99999)
```

```python
fx1(4.00001)
```

```python
fig, ax = plt.subplots()
plt.xlim(3.99999, 4.00001)
plt.ylim(5, 7)
_=ax.plot(x1,y1)
```

so as you can see both from left hand side and right hand side as x approaches 4 y getting closer to 6.

**The limit of the expression as x approaches 4 is therefore equal to 6.**
