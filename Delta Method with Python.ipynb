### Computing Derrivatives

## Delta Method   $\Delta x$


# $\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$

### here we work with this $ y = x^2 + 5x + 5 $ function 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    y = x**4 + 5*x + 5
    return y
```

```python
x = np.linspace(-10, 10, 1000)
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

### Goal Identify the slope where, x = 5 ###

```python
f(7)
```

#### Let's call this point A,  A is located at (7, 2441)

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='Lightgray')
plt.axhline(y=0, color='Lightgray')
plt.scatter(7, 2441)
_ = ax.plot(x,y)
```

#### The delta mathod uses the difference between two points to calculate slope of the point. So let's define B then, where x = 9

```python
f(9)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='Lightgray')
plt.axhline(y=0, color='Lightgray')
plt.scatter(7, 2441)
plt.scatter(9, 6611, c = 'orange', zorder=3)
_ = ax.plot(x,y)
```

#### To find the slope m between points A and B we need to use formula below:

##  $ m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{6611 - 2441}{9 - 7} = \frac{4170}{2} = 2085 $

```python
m = (6611-2441)/(9-7)
m
```

```python
b = 2441-m*7
b
```

#### it means that line crosses y-axis at y = -12154

```python
line_1 = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(7, 2441)
plt.scatter(9, 6611, c='orange', zorder=3)
plt.ylim(-500, 11000)
plt.plot(x, line_1, c='orange')
_ = ax.plot(x,y)
```

#### Here is the slope, but as much we move B closer to A as much we get closer to true tangent of point A.

let's now make B equal to 7.2

### $ \Delta x = x_2 - x_1 = 7.2 - 7.0 = 0.2  $

```python
f(7.2)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(7, 2441)
plt.scatter(7.2, 2728.3856, c = 'orange', zorder=3)
_ = ax.plot(x,y)
```

```python
m = (2728.3856 - 2441)/(7.2-7)
m
```

```python
b = 2728.3856 - m*7.2
b
```

```python
line_2 = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(7, 2441)
plt.scatter(7.2, 2728.3856, c='orange', zorder=3)
plt.ylim(-500, 11000)
plt.plot(x, line_2, c='orange', zorder=3)
_ = ax.plot(x,y)
```

#### As we get closer to $ \Delta x $ approaches $ 0 $ we know better what the slope is for now I think it's close to 1400  let's do one more to see it.
